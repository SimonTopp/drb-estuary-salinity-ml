### Preprocess the data

```python
#Make initial plots of the data
plt.subplot(2,2,1)
plt.plot(data.nitrateobs)
plt.ylabel('Nitrate')

plt.subplot(2,2,2)
plt.plot(data.discharge)
plt.ylabel('Discharge')

plt.subplot(2,2,3)
plt.plot(data.prcp)
plt.ylabel('Precipitation')

plt.subplot(2,2,4)
plt.plot(data.tmin, label = 'tmin')
plt.plot(data.tmax, label = 'tmax')
plt.ylabel('Temperature')
plt.legend()

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,wspace=0.35)
```

>Remove the seasonal component of the signal and standardize each time series. This step will likely need to be revisited


```python
#function for removing seasonal signal
def remove_seasonal_signal(raw_data,j):
    seasonal_removed = list()
    for i in range(len(raw_data)):
        doy = raw_data.loc[i,'julianday']
        doy_mean = np.nanmean(raw_data[raw_data['julianday'] == doy].iloc[:,j])
        value = raw_data.iloc[i,j]-doy_mean
        seasonal_removed.append(value)
        seasonal_removed_np = np.array(seasonal_removed)
    return seasonal_removed_np

#function for standardizing the data
def standardize_data(sr):
    standardized = list()
    for i in range(len(sr)):
        value_z = (sr[i] - np.nanmean(sr))/np.nanstd(sr, ddof = 1)
        #value_z = (sr[i]-np.nanmin(sr))/(np.nanmax(sr)-np.nanmin(sr))
        standardized.append(value_z)
    return standardized

```

>The goal is to convert each variable to approximately normal distributions. View how this changes the distribution of data for variables (change the j value to view)

```python
#j = 2 -> nitrate observed; j = 3 -> nitrate modeled model A; j = 4 -> nitrate modeled model B
#j = 5 -> discharge; j = 6 -> precip; j = 7 -> tmax; j = 8 -> tmin
j = 8
#data
name = data.columns.values[j]
seasonal_removed_data = remove_seasonal_signal(data,j)
standarized_data = standardize_data(sr = seasonal_removed_data)

plt.subplot(3,1,1)
plt.hist(data.iloc[:,j])
plt.title(name + ' Raw')

plt.subplot(3,1,2)
plt.hist(seasonal_removed_data)
plt.title(name + ' Seasonal Signal Removed')

plt.subplot(3,1,3)
plt.hist(standarized_data)
plt.title(name + ' Standardized')
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,wspace=0.35)
plt.tight_layout()
```

>Create the matrix for calculating information theory metrics

```python
#targets
#remove seasonal signal
ssn_rmv_nobs, ssn_rmv_mdla, ssn_rmv_mdlb = remove_seasonal_signal(data,2), remove_seasonal_signal(data,3), remove_seasonal_signal(data,4)
#standardize
std_nobs, std_mdla, std_mdlb = standardize_data(ssn_rmv_nobs), standardize_data(ssn_rmv_mdla), standardize_data(ssn_rmv_mdlb)

#sources
#remove seasonal signal -- discharge and precip
ssn_rmv_discharge, ssn_rmv_precip = remove_seasonal_signal(data,5), remove_seasonal_signal(data,6)
#remove seasonal signal -- tmin and tmax
ssn_rmv_tmin, ssn_rmv_tmax = remove_seasonal_signal(data,7), remove_seasonal_signal(data,8)
#standardize -- discharge and precip
std_discharge, std_precip = standardize_data(ssn_rmv_discharge), standardize_data(ssn_rmv_precip)
#standardize -- tmin and tmax
std_tmin, std_tmax = standardize_data(ssn_rmv_tmin), standardize_data(ssn_rmv_tmax)

# create data frame
d = {'nobs': std_nobs, 'nmdla': std_mdla, 'nmdlb': std_mdlb, 
    'discharge': std_discharge, 'precip': std_precip, 'tmin': std_tmin, 'tmax': std_tmax}

d_small = {'discharge': std_discharge, 'precip': std_precip}

data_prepped = pd.DataFrame(data = d)
```

### Calculate Mutual Information

>*mutual information* - the amount of information obtained by one variable when observing another. Here mutual information between source and sink variables is normalized by the shannon entropy of the sink variables, so that the output can be conceptualized as the fraction of entropy in the sink variable that can be explained by the source variable. __[More info](https://github.com/pdirmeyer/l-a-cheat-sheets/blob/main/Coupling_metrics_V30_MI.pdf)__

>Additionally, the critical mutual information threshold is calculated by shuffling the time series to destroy temporal relationships and calculating the mutual information 1000 times (numiter variable)

```python
#choose a source variable and view the mututal information between it and the model runs
source = 'tmax'
sinks = ['nobs','nmdla','nmdlb']
MI_mat = np.empty([0,1])
MI_crit_mat = np.empty([0,1])
bins  = [11,11,11]
for snk in sinks:
    paired_for_calcs = data_prepped[[source,snk]].to_numpy()
    #print(paired_for_calcs)
    MI, n = TEpython_ParallelNAN2.mutinfo_new(paired_for_calcs, nbins = bins)
    MIcrit = TEpython_ParallelNAN2.mutinfo_crit_newPar(paired_for_calcs, nbins = bins,  alpha = 0.05, numiter = 1000 ,ncores = 8)
    MI_mat = np.append(MI_mat,MI)
    MI_crit_mat = np.append(MI_crit_mat, MIcrit[0])
MI_mat, MI_crit_mat
```

### Plot mutual information results

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
nitrate = ['Observed','Model A','Model B']
mi = MI_mat
barlist = ax.bar(nitrate,mi)
barlist[0].set_color('gray')
barlist[1].set_color('blue')
barlist[2].set_color('red')
plt.plot([-.45, .45], [MI_crit_mat[0], MI_crit_mat[0]], '-o', c = 'black')
plt.plot([0.55,1.45], [MI_crit_mat[1], MI_crit_mat[1]], '-o', c = 'black')
plt.plot([1.55,2.45], [MI_crit_mat[2], MI_crit_mat[2]], '-o', c = 'black')
plt.title('Mutual Information Between '+source+' and nitrate')
plt.ylabel('Mutual information (bits/bits)')
plt.show()
```

>Choose 'tmax' as the source variable and view how both the observations (gray) and model B (red) show similar mutual information between tmax and nitrate (about 0.07) compared to model A (blue), which shows lower shared information between tmax and modeled nitrate. Critical mutual information threshold is shown with black lines.

>Model A shows significantly less coupling between tmax and nitrate, which could be one of the reasons that the results are worse. The same is true for discharge, and precipitation to a lesser extent.

### Calculate Transfer Entropy

>*transfer entropy* - the reduction of uncertainty in the sink variable (Y) due to knowledge of the source variable (X) independent of reduction in uncertainty in the sink variable due to knowledge of its own past. The calculation is done over a range of time lags from 0 to max_lag. Here mutual information between source and sink variables is normalized by the shannon entropy of the sink variables, so that the output can be conceptualized as the fraction of entropy in the sink variable that can be explained by the source variable at a certain time lag, in excess of the sink variable's own history. __[More info](https://github.com/pdirmeyer/l-a-cheat-sheets/blob/main/Coupling_metrics_V30_TE.pdf)__

>Additionally, the critical transfer entropy threshold is calculated at each time lag by shuffling the time series to destroy temporal relationships and calculating the transfer entropy 1000 times (numiter variable)


```python
source = 'tmax'
sinks = ['nobs','nmdla','nmdlb']

max_lag = 180

te_list = []
te_crit_list = []

for k in range(len(sinks)):
    paired_for_calcs = data_prepped[[source,sinks[k]]].to_numpy()
    te_list.append([])
    te_crit_list.append([])
    print(source, '-->', sinks[k])
    for i in range(max_lag):
        T, N, M4short = TEpython_ParallelNAN2.transen_new2(paired_for_calcs, shift = [i,0,i], nbins = bins)
        Tcrit = TEpython_ParallelNAN2.transen_crit_new2(paired_for_calcs, shift = [i,0,i], alpha = 0.5, numiter = 1000, nbins = bins, ncores = 8)
        te_list[k].append(T)
        te_crit_list[k].append(Tcrit)
        
       
```

### Plot Results

```python
#plot the data
#observed
plt.plot(te_list[0],'-', c='gray', label = 'Observed')
plt.plot(te_crit_list[0], '--', c='gray', lw = 0.5, label = 'Obs Crit')
#model A
plt.plot(te_list[1],'-', c='blue', label = 'Model A')
plt.plot(te_crit_list[1], '--', c='blue', lw = 0.5, label = 'Model A Crit')
#model B
plt.plot(te_list[2],'-', c='red', label = 'Model B')
plt.plot(te_crit_list[2], '--', c='red', lw = 0.5, label = 'Model B Crit')
plt.title('Transfer Entropy -- source: '+source)
plt.ylabel('TE sink to source')
plt.xlabel('Time Lag (days)')
plt.legend()
```

> Choose 'tmax' as the source variable and view how Model A (red) and the Observations (gray) show good agreement across a range of time lags, while Model B (blue) deviates particularly after ~10 days. 

```python
#plot the data zoomed in 
#observed
plt.plot(te_list[0],'-', c='gray', label = 'Observed')
plt.plot(te_crit_list[0], '--', c='gray', lw = 0.5, label = 'Obs Crit')
#model A
plt.plot(te_list[1],'-', c='blue', label = 'Model A')
plt.plot(te_crit_list[1], '--', c='blue', lw = 0.5, label = 'Model A Crit')
#model B
plt.plot(te_list[2],'-', c='red', label = 'Model B')
plt.plot(te_crit_list[2], '--', c='red', lw = 0.5, label = 'Model B Crit')
plt.title('Transfer Entropy -- source: '+source)
plt.ylabel('TE sink to source')
plt.xlabel('Time Lag (days)')
plt.xlim([0,25])
plt.legend()
```

>Zoom in on time lags from 0 to 10 days. Model B and Observed show good agreement with the peak of TE at a lag of 3 days (above the TE critical threshold), while Model A misses this completely.

#### Concluding thoughts

>- The analysis of mutual information between inputs and modeled outputs can tell us how subsystems are coupled within models and how that coupling has changed due to modeling decisions
>- The comparison of mutual information of modeled results to observations tells us if the models are overdetermined, or if they are underutilizing the information present in the input time series 
>- The analysis of transfer entropy allows us to examine the one-way transfer of information from input variables to output, the comparison with observed data allows us to determine how faithfully the models are representing the driving processes
>- TE analysis also allows us to observe the critical time lags associated with these processes, and how modeling decisions have affected the timing of drivers within the model
>- For example, maximum temperature should signficantly reduce variance in nitrate at the 3 day time lag. In Model A this is not captured, however in Model B this dynamic is well captured
>- This analysis and interpretation has been fairly qualitative, but I can imagine an analysis of the TE and MI results that gets really quantitative by looking at the time lags where T > Tcrit and comparing those between model runs. Also there has been some interesting work done by __[Ruddell et al.](https://agupubs.onlinelibrary.wiley.com/doi/full/10.1029/2018WR023692)__ looking at MI and TE together and what it tells you about a model
